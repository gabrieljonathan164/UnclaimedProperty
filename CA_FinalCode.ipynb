```python
import pandas as pd
import numpy as np
from math import radians
from math import dist
import matplotlib as plt
import re
from pandas_profiling import ProfileReport
```

```python
#Import Datasets

df1 = pd.read_csv('CA1.csv', encoding='latin1', low_memory = False)
df2 = pd.read_csv('CA2.csv', encoding='latin1', low_memory = False)
df3 = pd.read_csv('CA3.csv', encoding='latin1', low_memory = False)
```

```python
#Merge the datasets

result = [df1, df2, df3]
df_CA = pd.concat(result)
```

```python
df_CA
```

```python
#Extracting a profile report Before Data wrangling for the California UCP DataFrame to understand the information under different variable.
profile_CA = ProfileReport(df_CA, title = "California UCP Profile Report Before Data wrangling ")
profile_CA.to_file("CA_profile_report_Before_Wrangling.html")
```

```python
df_CA.columns
```

```python
#Cleaning the null values.

df_CA['OWNER_NAME'].fillna("No Name", inplace = True)
df_CA['OWNER_STREET_1'].fillna("NaN", inplace = True)
df_CA['OWNER_STREET_2'].fillna(" ", inplace = True)
df_CA['OWNER_STREET_3'].fillna(" ", inplace = True)
df_CA['OWNER_CITY'].fillna("No City", inplace = True)
df_CA.loc[df_CA['OWNER_CITY'].str.contains("ZZ"), 'OWNER_CITY'] = "No City"

df_CA
```

```python
#Created a new column Owner's address by merging 'OWNER_STREET_1', 'OWNER_STREET_2', 'OWNER_STREET_3'
df_CA.insert(loc = 8, column = 'ADDRESS', value = df_CA['OWNER_STREET_1'].map(str) + ' ' +  df_CA['OWNER_STREET_2'].map(str) + ' ' +  df_CA['OWNER_STREET_3'].map(str))
df_CA['ADDRESS']
```

```python
#Cleaning the Address
df_CA.loc[df_CA['ADDRESS'].str.contains("NaN|ZZ"), 'ADDRESS'] = "No ADDRESS"

df_CA
```

```python
#Subsetting with the required columns
df_CA_subset = df_CA[['PROPERTY_ID', 'PROPERTY_TYPE', 'DATE_REPORTED','OWNER_NAME', 'ADDRESS','OWNER_CITY','OWNER_ZIP', 'CURRENT_CASH_BALANCE', 'HOLDER_NAME']]
df_CA_subset
```

```python
#Data Formatting - Rename the columns 
df_CA_subset.rename(columns={'OWNER_CITY': 'CITY','OWNER_ZIP':'ZIPCODE', 'CURRENT_CASH_BALANCE': 'UNCLAIMED MONEY'}, inplace=True)
df_CA_subset
```

```python
#Cleaning Zipcodes

df_CA_subset['ZIPCODE'] = df_CA_subset['ZIPCODE'].str.replace(' |-','')
df_CA_subset['ZIPCODE'] = df_CA_subset['ZIPCODE'].str.slice(stop=5)

df_CA_subset['ZIPCODE'].fillna("No Zipcode", inplace = True)
df_CA_subset.loc[df_CA_subset['ZIPCODE'].str.contains("00000"), 'ZIPCODE']="No Zipcode"
df_CA_subset.loc[df_CA_subset['ZIPCODE'].str.contains('^\s*$'), 'ZIPCODE']="No Zipcode"

df_CA_subset['ZIPCODE']
```

```python
#Subset with only JPMORGAN CHASE BANK , BANK OF AMERICA , WELLS FARGO , BANK OF NEW YORK MELLON CORPORATE TRUST 
df_CA_Banksubset = df_CA_subset[df_CA_subset['HOLDER_NAME'].str.contains("JPMORGAN CHASE BANK|BANK OF AMERICA|WELLS FARGO|BANK OF NEW YORK MELLON CORPORATE TRUST")]
df_CA_Banksubset
```

```python
#Cleaning "BANK OF AMERICA"
df_CA_Banksubset.loc[df_CA_Banksubset['HOLDER_NAME'].str.contains("BANK OF AMERICA"), 'HOLDER_NAME']="BANK OF AMERICA"

#Cleaning "WELLS FARGO BANK N A"
df_CA_Banksubset.loc[df_CA_Banksubset['HOLDER_NAME'].str.contains("WELLS FARGO"), 'HOLDER_NAME']="WELLS FARGO BANK N A"

#Cleaning "JPMORGAN CHASE BANK"
df_CA_Banksubset.loc[df_CA_Banksubset['HOLDER_NAME'].str.contains("JPMORGAN CHASE BANK"), 'HOLDER_NAME']="JPMORGAN CHASE BANK"

#Cleaning "BANK OF NEW YORK MELLON CORPORATE TRUST"
df_CA_Banksubset.loc[df_CA_Banksubset['HOLDER_NAME'].str.contains("BANK OF NEW YORK MELLON CORPORATE TRUST"), 'HOLDER_NAME']="BANK OF NEW YORK MELLON CORPORATE TRUST"

df_CA_Banksubset
```

```python
#Remove duplicates if available
df_CA_Banksubset.drop_duplicates()
df_CA_Banksubset
```

```python
#Check for missing values
for x in df_CA_Banksubset.columns:
    missing_val= df_CA_Banksubset[x].isnull().sum()
 
    if missing_val>0:
        print("{} has {} missing value(s)".format(x,missing_val))
    else:
        print("{} has NO missing value!".format(x))
```

```python
Output_df_CA = df_CA_Banksubset.sort_values(by='PROPERTY_ID')
Output_df_CA
```

```python
#Extracting a profile report After Data wrangling for the California UCP DataFrame to understand the information under different variable.
profile_CA = ProfileReport(Output_df_CA, title = "California UCP Profile Report After Data wrangling ")
profile_CA.to_file("CA_profile_report_After_Wrangling.html")
```
